## 1. `plt.plot(x, y)` — Line Chart

Draws a line chart connecting your data points — the go-to choice for **trends over time**.

**Key parameters:**

| Parameter | Description |
|---|---|
| `x, y` | Sequences of numbers — `x` is the horizontal axis, `y` is the vertical axis |
| `color=` | Line color — name (`"red"`) or hex code (`"#00D4AA"`) |
| `linewidth=` | Thickness of the line, in points |
| `marker=` | Symbol drawn at each data point: `'o'`, `'s'`, `'^'` ... |


```python
months = [1, 2, 3, 4, 5, 6]
sales  = [10, 15, 13, 22, 18, 28]

plt.plot(months, sales,
         color="#00D4AA",
         linewidth=2.5,
         marker="o")

plt.title("Monthly Sales", color='#00D4AA')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

weeks = [1, 2, 3, 4, 5, 6]
subscribers = [50, 85, 70, 120, 105, 150]

plt.plot(weeks, subscribers,
         color="#0055ff", 
         linewidth=2.3,
         marker="o")

plt.title("Weeks and Subscribers", color="#0055ff")
plt.xlabel("Weeks")
plt.ylabel("Subscribers")

plt.show()
```

**Notes**
- Use `plt.plot()` for continuous data, trends, and comparing multiple series.
- Call `plt.plot()` again *before* `plt.show()` to add another line to the same chart.
- Add `plt.legend()` when you have more than one line, so viewers know which is which.


```python
# Multiple lines on the same chart
temperature = [18, 19, 21, 25, 23, 27]
rainfall_mm = [10, 8, 6, 3, 5, 2]

plt.plot(months, temperature, color="#FF6B35", linewidth=2.5, marker="o", label="Temperature (°C)")
plt.plot(months, rainfall_mm, color="#22D3EE", linewidth=2.5, marker="s", label="Rainfall (mm)")

plt.title("Temperature vs Rainfall")
plt.xlabel("Month")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


weeks = [1, 2, 3, 4, 5, 6]


subscribers_youtube = [50, 85, 70, 120, 105, 150]
subscribers_tiktok = [30, 60, 95, 110, 140, 180]


plt.plot(weeks, subscribers_youtube, color="#FF6B35", linewidth=2.5, marker="o", label="YouTube")

plt.plot(weeks, subscribers_tiktok, color="#22D3EE", linewidth=2.5, marker="s", label="TikTok")

plt.title("YouTube vs TikTok Subscribers")
plt.xlabel("Weeks") 
plt.ylabel("New Subscribers")

plt.legend() 
plt.show()
```

## 2. `plt.bar(x, y)` — Bar Chart

Draws a bar chart — best for **comparing values across categories**.

**Key parameters:**

| Parameter | Description |
|---|---|
| `x, y` | Category labels (or positions) and their corresponding values |
| `color=` | Fill color of the bars — one color, or a list for per-bar colors |
| `width=` | Bar thickness (default `0.8`). Smaller values add gaps between bars |
| `plt.barh()` | Same idea, but draws horizontal bars — handy for long category names |


```python
subjects = ["Math", "Sci", "Art", "PE"]
grades   = [85, 92, 78, 95]

plt.bar(subjects, grades, color="#FF6B35")

plt.title("Grade Comparison")
plt.xlabel("Subject")
plt.ylabel("Grade")
plt.show()
```

**Notes**
- Use `plt.bar()` for comparing discrete categories — subjects, products, regions, survey answers.
- It is *not* the right tool for continuous trends over time — use `plt.plot()` for that instead.
- Sort bars by value, or color the highest bar differently, to draw the eye.


```python
# Horizontal bars — handy for long category names
subjects_long = ["Mathematics", "Computer Science", "Fine Arts", "Physical Education"]

plt.barh(subjects_long, grades, color="#00D4AA")
plt.title("Grade Comparison (horizontal)")
plt.xlabel("Grade")
plt.show()
```

## 3. `plot()` vs `bar()` — which one do I use?

| | `plt.plot()` | `plt.bar()` |
|---|---|---|
| Best for | Continuous data & trends | Discrete, categorical data |
| X-axis | Time or an ordered sequence | Separate, unordered-ish categories |
| Emphasizes | Shape / direction of change | Size / magnitude of each category |
| Multiple series | Several lines on one chart | Bars side-by-side or stacked |
| Examples | Stock price, temperature over a year, website traffic | Sales by region, votes by candidate, grades by subject |

**Rule of thumb:** if you can meaningfully connect the dots with a line, use `plot()`. If the x-axis is a list of names, use `bar()`.

## 4. Quick Reference

Functions you'll use alongside `plt.plot()` and `plt.bar()`:

| Function | Purpose |
|---|---|
| `plt.title()` | Main chart title |
| `plt.xlabel()` / `plt.ylabel()` | X and Y axis labels |
| `plt.legend()` | Show labels for each dataset |
| `color=` / `linewidth=` | Style the line or bar |
| `marker=` `'o'` `'s'` `'^'` | Data point markers on line charts |
| `plt.figure(figsize=)` | Set chart width × height in inches |
| `plt.savefig('f.png')` | Save chart to image file |
| `plt.subplot(r, c, i)` | Place multiple charts in one figure |
| `plt.tight_layout()` | Auto-fix spacing between subplots |
| `plt.show()` | Display chart in window/notebook |


```python
# plt.subplot() — placing a line chart and a bar chart side by side
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(months, sales, color="#00D4AA", linewidth=2.5, marker="o")
plt.title("Monthly Sales (line)")
plt.xlabel("Month")

plt.subplot(1, 2, 2)
plt.bar(subjects, grades, color="#FF6B35")
plt.title("Grade Comparison (bar)")
plt.xlabel("Subject")

plt.tight_layout()
plt.show()
```

## 5. Seaborn — Beautiful Statistical Visualizations

Seaborn is built on top of Matplotlib and works directly with **Pandas DataFrames**. It adds statistical charts and much nicer default styling.

Topics: `barplot` · `boxplot` · `heatmap` · `pairplot` · `themes` · `hue`


```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid")  # a Seaborn "theme"

# Sample dataset
np.random.seed(42)
df = pd.DataFrame({
    "subject": ["Math", "Sci", "Art", "PE"] * 20,
    "grade": np.concatenate([
        np.random.normal(85, 5, 20),
        np.random.normal(90, 6, 20),
        np.random.normal(78, 8, 20),
        np.random.normal(93, 4, 20),
    ]),
    "term": np.random.choice(["Term 1", "Term 2"], 80)
})
df.head()
```

### `sns.barplot()`
Like `plt.bar()`, but automatically aggregates repeated observations and adds confidence-interval error bars.


```python
plt.figure(figsize=(7, 4.5))
sns.barplot(data=df, x="subject", y="grade", hue="term", palette=["#00D4AA", "#FF6B35"])
plt.title("Average Grade by Subject and Term")
plt.show()
```

### `sns.boxplot()`
Shows the distribution (median, quartiles, outliers) of a numeric variable per category — great for spotting spread and outliers.


```python
plt.figure(figsize=(7, 4.5))
sns.boxplot(data=df, x="subject", y="grade", palette="Set2")
plt.title("Grade Distribution by Subject")
plt.show()
```

### `sns.heatmap()`
Visualizes a matrix of values (often a correlation matrix) as a grid of colors — perfect for spotting relationships between numeric columns at a glance.


```python
# Correlation heatmap
scores = pd.DataFrame(np.random.rand(50, 4), columns=["Math", "Science", "Art", "PE"])
scores["Science"] = scores["Math"] * 0.8 + np.random.rand(50) * 0.2  # correlate two columns

plt.figure(figsize=(6, 5))
sns.heatmap(scores.corr(), annot=True, cmap="magma", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

**Interpretation:** values close to **1** mean two subjects' scores rise together; values close to **-1** mean one rises as the other falls; values near **0** mean little relationship. Notice `Math` and `Science` are strongly correlated by construction above.


### `sns.pairplot()`
Plots every numeric column against every other column in a grid — a fast way to scan a whole dataset for relationships.


```python
sns.pairplot(scores, corner=True)
plt.suptitle("Pairplot of Subject Scores", y=1.02)
plt.show()
```

### Themes and `hue`
- `sns.set_theme(style=...)` — `"whitegrid"`, `"darkgrid"`, `"white"`, `"dark"`, `"ticks"`.
- `hue=` — splits a chart into colored sub-groups by a categorical column (used above to separate `Term 1` / `Term 2`).


```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(7, 4.5))
sns.boxplot(data=df, x="subject", y="grade", hue="term", palette="viridis")
plt.title("Grades by Subject and Term (darkgrid theme)")
plt.show()

sns.set_theme(style="whitegrid")  # reset for later cells
```

## 6. Plotly — Interactive Charts

Plotly Express (`px`) builds **interactive** charts: hover tooltips, zoom, pan, and export — all in the browser/notebook, no extra code needed.

Topics: what makes Plotly different · `px.bar()`, `px.line()`, `px.scatter()` · `hover_data` · `px.pie()` and `px.sunburst()` · exporting as HTML · Plotly vs Matplotlib vs Seaborn


### 1. What makes Plotly different
- Charts are **interactive**: hover for tooltips, scroll/drag to zoom and pan, click legend items to toggle series.
- Works great in notebooks and can be exported as a standalone HTML file or embedded in a web page.
- Same "grammar" as Seaborn: pass a DataFrame and column names.


```python
import plotly.express as px

sales_df = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "sales": [10, 15, 13, 22, 18, 28],
    "city": ["Cairo", "Cairo", "Cairo", "Cairo", "Cairo", "Cairo"]
})
sales_df
```

### 2. `px.bar()`, `px.line()`, `px.scatter()`
Same idea as `plt.bar()` / `plt.plot()`, but interactive.


```python
fig = px.bar(sales_df, x="month", y="sales", title="Monthly Sales", color_discrete_sequence=["#8B5CF6"])
fig.show()
```

```python
fig = px.line(sales_df, x="month", y="sales", markers=True, title="Monthly Sales Trend",
              color_discrete_sequence=["#00D4AA"])
fig.show()
```

```python
scatter_df = pd.DataFrame({
    "study_hours": np.random.uniform(1, 10, 40),
})
scatter_df["grade"] = 60 + scatter_df["study_hours"] * 3 + np.random.normal(0, 5, 40)

fig = px.scatter(scatter_df, x="study_hours", y="grade", title="Study Hours vs Grade",
                  color_discrete_sequence=["#FF6B35"])
fig.show()
```

### 3. Adding `hover_data`
Pass extra columns to `hover_data=[...]` to show more info in the tooltip without cluttering the chart itself.


```python
fig = px.bar(sales_df, x="month", y="sales", hover_data=["city"],
             title="Monthly Sales (hover to see city)",
             color_discrete_sequence=["#22D3EE"])
fig.show()
```

### 4. `px.pie()` and `px.sunburst()`
- `px.pie()` — parts of a whole, one level of categories.
- `px.sunburst()` — hierarchical parts of a whole (categories nested inside categories), shown as concentric rings.


```python
fig = px.pie(sales_df, names="month", values="sales", title="Sales Share by Month",
             color_discrete_sequence=px.colors.sequential.Purp)
fig.show()
```

```python
region_df = pd.DataFrame({
    "region": ["North", "North", "South", "South", "East", "East"],
    "product": ["Phones", "Laptops", "Phones", "Laptops", "Phones", "Laptops"],
    "revenue": [120, 80, 90, 60, 70, 50]
})

fig = px.sunburst(region_df, path=["region", "product"], values="revenue",
                   title="Revenue by Region and Product")
fig.show()
```

### 5. Exporting as HTML
Save any Plotly figure as a standalone, interactive HTML file that opens in any browser — no Python needed to view it.


```python
fig = px.line(sales_df, x="month", y="sales", markers=True, title="Monthly Sales Trend")
fig.write_html("monthly_sales.html")
print("Saved to monthly_sales.html")
```

### 6. Plotly vs Matplotlib vs Seaborn

| | Matplotlib | Seaborn | Plotly |
|---|---|---|---|
| Interactivity | Static | Static | **Interactive** (hover, zoom, pan) |
| Best for | Full control, publication figures | Quick statistical charts from DataFrames | Dashboards, exploration, sharing on the web |
| Input | Lists / arrays | Pandas DataFrame | Pandas DataFrame |
| Output | PNG/PDF/SVG | PNG/PDF/SVG (via Matplotlib) | Interactive HTML / notebook widget |
| Learning curve | Steepest (most manual) | Easy, sensible defaults | Easy, but heavier files |

**Rule of thumb:** start in Matplotlib for full control, reach for Seaborn when your data is in a DataFrame and you want a quick statistical view, and use Plotly when the chart needs to be interactive or shared as a webpage.

# Seaborn — Beautiful Statistical Visualizations

**Level 3 · Lecture 2**

Seaborn is built on top of Matplotlib and works hand-in-hand with `pandas` DataFrames. It adds statistical smarts and much better default styling.

We'll cover:
- `sns.barplot()`
- `sns.boxplot()`
- `sns.heatmap()`
- `sns.pairplot()`
- **Themes** (`sns.set_theme()`)
- **`hue=`** — coloring by a category

**Task:** Correlation Heatmap + Interpretation


```python
import seaborn as sns
import pandas as pd

# A classic built-in dataset: restaurant tips
tips = sns.load_dataset("tips")
tips.head()
```

## `sns.set_theme()` — Themes

Call this once at the top of your notebook to switch every seaborn (and matplotlib) chart to a polished style.

Built-in styles: `"darkgrid"`, `"whitegrid"`, `"dark"`, `"white"`, `"ticks"`.


```python
sns.set_theme(style="whitegrid", palette="muted")
```

## `sns.barplot()`

Like `plt.bar()`, but works directly with column names from a DataFrame and can automatically compute a summary statistic (mean, by default) with a confidence-interval error bar.


```python
plt.figure(figsize=(7, 4.5))
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Average Bill by Day and Gender")
plt.show()
```

**`hue=`** splits each category into sub-groups and colors them differently — here, Male vs Female per day. This is the fastest way to add a third dimension to almost any seaborn chart.


## `sns.boxplot()`

Shows the distribution of a numeric variable (median, quartiles, outliers) per category — great for spotting spread and outliers that a bar chart hides.


```python
plt.figure(figsize=(7, 4.5))
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.title("Bill Distribution by Day and Smoker Status")
plt.show()
```

## `sns.pairplot()`

Plots every numeric column against every other numeric column — a fast way to scan an entire dataset for relationships in one call.


```python
sns.pairplot(tips, vars=["total_bill", "tip", "size"], hue="time")
plt.show()
```

## `sns.heatmap()`

Visualizes a matrix of values (very often a correlation matrix) as a grid of colored cells — perfect for spotting which numeric columns move together.


```python
numeric_cols = tips.select_dtypes(include="number")
corr = numeric_cols.corr()

plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Correlation Matrix — Tips Dataset")
plt.show()
```

# Plotly — Interactive Charts

**Level 3 · Lecture 4**

### What makes Plotly different
Unlike Matplotlib/Seaborn (static images), Plotly charts are **interactive**: hover for tooltips, zoom, pan, and export — right inside the notebook or a web page.

We'll cover:
1. What makes Plotly different
2. `px.bar()`, `px.line()`, `px.scatter()`
3. Adding `hover_data`
4. `px.pie()` and `px.sunburst()`
5. Exporting as HTML
6. Plotly vs Matplotlib vs Seaborn


```python
import plotly.express as px
import plotly.io as pio

# Use a renderer that embeds the interactive chart as portable HTML+JS
# (renders reliably across Jupyter, JupyterLab, and VS Code)
pio.renderers.default = "notebook_connected"

# Reuse the tips DataFrame from the Seaborn section
tips.head()
```

## `px.bar()`, `px.line()`, `px.scatter()`

The same three chart types we've already met — now interactive. The API mirrors seaborn: pass a DataFrame and column names.


```python
fig = px.bar(tips, x="day", y="total_bill", color="sex", barmode="group",
             title="Average Bill by Day and Gender")
fig.show()
```

```python
daily_avg = tips.groupby("day", as_index=False, observed=True)["total_bill"].mean()

fig = px.line(daily_avg, x="day", y="total_bill", markers=True,
              title="Average Bill by Day (Line)")
fig.show()
```

```python
fig = px.scatter(tips, x="total_bill", y="tip", color="time", size="size",
                  title="Tip vs Total Bill")
fig.show()
```

## Adding `hover_data`

Pass a list of extra columns to `hover_data=` to show more context when a user hovers over a point or bar — no extra chart needed.


```python
fig = px.scatter(tips, x="total_bill", y="tip", color="day",
                  hover_data=["sex", "smoker", "size"],
                  title="Tip vs Total Bill (with extra hover info)")
fig.show()
```

## `px.pie()` and `px.sunburst()`

`px.pie()` shows parts of a whole. `px.sunburst()` extends that idea to **hierarchical** categories — rings inside rings.


```python
fig = px.pie(tips, names="day", values="total_bill", title="Share of Total Bill by Day")
fig.show()
```

```python
fig = px.sunburst(tips, path=["day", "time", "sex"], values="total_bill",
                   title="Total Bill by Day → Time → Gender")
fig.show()
```

## Exporting as HTML

A Plotly figure can be saved as a **standalone interactive HTML file** — no Python needed to view it, just a browser.


```python
fig.write_html("sunburst_chart.html")
print("Saved sunburst_chart.html — open it in any browser.")
```

## Plotly vs Matplotlib vs Seaborn

| | Matplotlib | Seaborn | Plotly |
|---|---|---|---|
| Output | Static image | Static image | Interactive (hover/zoom/pan) |
| Best for | Full control, publication figures | Quick statistical charts from DataFrames | Dashboards, exploration, web sharing |
| Learning curve | Steeper (low-level) | Gentle (built on Matplotlib) | Gentle (similar API to Seaborn) |
| Export | PNG, PDF, SVG via `savefig()` | Same as Matplotlib | HTML, PNG (with `kaleido`), embeddable in web apps |

**Rule of thumb:** prototype fast with Seaborn, reach for Plotly when the audience needs to interact with the chart themselves, and drop to raw Matplotlib when you need pixel-level control.
